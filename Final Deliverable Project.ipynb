```python
# Function to conduct the ADF test and print results
def adf_test(series, series_name):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Test Results for {series_name}:')
    print(f'Test Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
```

```python
# ADF test for spot prices in levels
adf_test(df_spot, 'Spot Price Levels')
```

```python
# ADF test for spot prices first differences
adf_test(df_spot.diff().dropna(), 'Spot Price First Differences')
```

```python
# ADF test for log-spot prices levels
adf_test(np.log(df_spot), 'Log-Spot Price Levels')
```

```python
# ADF test for log-spot prices first differences
adf_test(np.log(df_spot).diff().dropna(), 'Log-Spot Price First Differences')
```

```python
def draw_correlated_normals(mn1 = 0.0, sd1 = 1.0, mn2 = 0.0, sd2 = 1.0, rho = 0.5, num_reps = 100):
  z1 = np.random.normal(size=num_reps, loc=mn1, scale=sd1)
  z2 = np.random.normal(size=num_reps, loc=mn2, scale=sd2)
  z2 = rho * z1 + np.sqrt((1.0 - rho**2.0)) * z2
  return (z1,z2)
```

```python
# Spot Price Dynamics from B & W
s_a1 = 0.342
s_b1 = 0.539
s_s1 = 0.110
s_S1 = 0.690
s_a2 = 0.391
s_b2 = 0.560
s_s2 = 0.116
s_S2 = 0.800
s_rho = 0.705

num_days = 45


ln_spot1 = np.zeros(num_days) ## Heating Oil
ln_spot2 = np.zeros(num_days) ## Gasoline


ln_spot1[0] = np.log(s_S1)
ln_spot2[0] = np.log(s_S2)


z1, z2 = draw_correlated_normals(sd1=s_s1, sd2=s_s2, rho=s_rho, num_reps=num_days)

for t in range(1, num_days):
  ln_spot1[t] = ln_spot1[t-1] + s_a1 * (s_b1 - np.exp(ln_spot1[t-1])) + z1[t]
  ln_spot2[t] = ln_spot2[t-1] + s_a2 * (s_b2 - np.exp(ln_spot2[t-1])) + z2[t]


ts = pd.DataFrame({'Heating Oil' : ln_spot1, 'Gasoline' : ln_spot2})
```
