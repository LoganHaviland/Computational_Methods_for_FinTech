**Q5.** Following Skyrms [Sky12] demonstrate the Dutch book argument for a Tootsie Roll economy with two flavors:
chocolate and cherry. What is the relationship between arbitrage in this simple economy and subjective probability?
Does this necessarily give us a derivation of Bayes’ Rule? Why or why not?


A5. The Dutch book argument for a Tootsie Roll economy would go as follows: We have two flavors (chocolate and cherry) Tootsie Rolls and individuals have their own individual subjective probability regarding some events that affect the price of either flavors. Constructing a Dutch book could be making a series of bets to these individuals about the probability of these events occuring and their related prices. If those series of bets are not coherent, the individual sets themselves up for a loss no matter what actually occurs (the events and resulting prices).

Arbitrage is similar in trying to make a risk free profit, the ideas hold true for both arbitrage and subjective probabilities creating a Dutch Book: a guarenteed profit/loss for the bet taker/subjective individual.

The way I understand is that the idea of arbitrage/Dutch Books do lead us into a better understanding of subjective probabilities and ideas related to that like coherence. Individuals have a prior belief about certain events and how that will change the price of the Cherry or Chocolate (to continue the analogy), and as we experience new information (the time periods increase) we continually update that belief with the new information. We then have our posterior distribution belief and that becomes our prior for the next period. Thus deriving the basic formula and idea of Bayes' Rule! Yes it does.

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import coint, adfuller
import io

file_name = 'WTI-Prices-1992-to-1993.xlsx'

# Put the file into a data frame
df = pd.read_excel(io.BytesIO(uploaded[file_name]))
# Show the file
#print(df)
```

```python
# Separate DataFrames for each column
df_date = df[['Date']].copy()
df_spot = df[['Spot']].copy()
df_futures = df[['Futures']].copy()
```

```python
#print(df_futures)
```

```python
# Function to conduct the ADF test and print results
def adf_test(series, series_name):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Test Results for {series_name}:')
    print(f'Test Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Critical Values: {result[4]}')
```

```python
# ADF test for spot prices in levels
adf_test(df_spot, 'Spot Price Levels')
```

```python
# ADF test for spot prices first differences
adf_test(df_spot.diff().dropna(), 'Spot Price First Differences')
```

```python
# ADF test for log-spot prices levels
adf_test(np.log(df_spot), 'Log-Spot Price Levels')
```

```python
# ADF test for log-spot prices first differences
adf_test(np.log(df_spot).diff().dropna(), 'Log-Spot Price First Differences')
```

```python
# ADF test for futures prices in levels
adf_test(df_futures, 'Futures Price Levels')
```

```python
# ADF test for futures prices first differences
adf_test(df_futures.diff().dropna(), 'Futures Price First Differences')
```

```python
# ADF test for log-futures prices levels
adf_test(np.log(df_futures), 'Log-Futures Price Levels')
```

```python
# ADF test for log-futures prices first differences
adf_test(np.log(df_futures).diff().dropna(), 'Log-Futures Price First Differences')
```

**Analysis**

After doing the ADF test for unit roots I found that the only series that we could reject the null hypothesis of of the presence of a unit-root were the first differences of the series. • Using linear regression calculate a rolling minimum-variance hedge ratio for each day of 1993 based on 60 prior
days observations. Plot a time-series of the hedge-ratios. Comment on the nature of the plotted series. What does
this say about the dynamic stability of the hedge ratio?

```python
from statsmodels.regression.rolling import RollingOLS
```

```python
# Rolling minimum-variance hedge ratio
prior_obs = 60
rols = RollingOLS.from_formula('Spot ~ Futures', data=df, window=prior_obs)
rols_results = rols.fit()

# The hedge ratios
hedge_ratios = rols_results.params['Futures']
```

```python
hedge_ratios.tail(366)
```

```python
mean = np.mean(hedge_ratios.tail(366))
```

```python
print(f"The average hedge ratio of the 1993 time frame using 60 day prior observations is  {mean:.4}!")
```

```python
# Plot the time-series of hedge ratios
plt.figure(figsize=(10, 6))
plt.plot(hedge_ratios, label='Rolling Hedge Ratio')
plt.title(f'Rolling Minimum-Variance Hedge Ratio ({prior_obs}-day window)')
plt.xlabel('Date')
plt.ylabel('Hedge Ratio')
plt.legend()
plt.show()
```
